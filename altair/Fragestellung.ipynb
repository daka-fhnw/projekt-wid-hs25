```python
import pandas as pd

data = pd.read_csv('./data/Gesamtdatensatz.csv')
data_april = data[(data['timestamp'] >= '2024-04-01') & (data['timestamp'] <= '2024-04-31')]
data_april
```

```python
pivot = pd.pivot_table(
    data,
    index=['location_id', 'location_name'],
    values=['child_pedestrians_count', 'adult_pedestrians_count', 'pedestrians_count'],
    aggfunc='sum',
    fill_value=0
).reset_index()
pivot['kinder_anteil'] = pivot['child_pedestrians_count'] / pivot['pedestrians_count']
pivot
```

```python
import altair as alt

chart = alt.Chart(pivot).mark_bar().encode(
    x=alt.X('location_name').axis(
        title='Gebietsnamen', 
        titleFontSize=14, 
        labelFontSize=12, 
        labelAngle=0
    ), 
    y=alt.Y('kinder_anteil').axis(
        title='Anteil Kinder',
        titleFontSize=14, 
        labelFontSize=12, 
        format='.0%'
    ),
    color=alt.Color('location_name', legend=None)
).properties(
    width=620,
    height=360,
    title=alt.TitleParams(
        text=["Im April 2024 waren an der Bahnhofstrasse Süd", "die meisten Kinder unterwegs."],
        fontSize=20
    )
)
chart
```

```python
#print(chart.to_json())

import json
# Convert to dict
spec = chart.to_dict()

with open("hausaufgabe.json", "w") as f:
    json.dump(spec, f, indent=2)
```
